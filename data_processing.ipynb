```python
#One useful/utility function for bar plotting
def dispAvgLabelOnBarTop(ax):
    for p in ax.patches:
        ax.annotate(format(p.get_height(), '.2f'), 
                (p.get_x() + ((p.get_width() - 0.5) / 2.), p.get_height()), 
                ha = 'center', va = 'center', 
                xytext = (0, 10), 
                textcoords = 'offset points')
```

###  Analysis of Job Involvement feature

What is the meaning of job involvement and how relevant is it for Employee hiring?

Assumption 1: Job involvement means how well an employee is involved in his/her job  
>If true, Job involvement should be highly correlated with performance rating provided by Employee Manager on employee, but Analysis of JobInvolvment vs EmployeeRating(visually observed below) doesn't seem to correlate.  
>Hence this assumption is **WRONG**
     
Assumption 2: Job involvement means how involved an employee has to be to perform this job role
      
>If true, there should be direct correlation between Job role and Job involvement, but Analysis of Job involvement vs Employee job role doesn't seem to indicate good correlation(observed visually below).   
>Hence this assumption is **WRONG**
      
Assumption 3: Job involvement indicates difficulty of job, higher difficulty requires higher involvement  

>If true, Job involvement should directly relate to hourly rate for that role, but Analysis of Job involvement vs Employee hourly rate doesn't seem to indicate good correlation(observed visually below). 
>Hence this assumption is WRONG
      
**Conclusion**: I have not been able to extract meaning from JobInvolvement feature, hence this feature will NOT be considered for ML modelling that will be used for employee hiring.

### Analysis of EmpJobLevel feature

What is the meaning of Employee Job level and how relevant is it for Employee hiring?

Assumption 1: Job level means position of the employee in the companys hierarchy tree, starting from CEO at the top to Fresher/Intern at the bottom.  

>Count of employees in job level 5(top level) is least(56) and bottom two level is max. So, this assumption is possibly RIGHT.
              
>Absolute Correlation of EmployeeJobLevel with total work experience in the below list indicates that this assumption is RIGHT.
              
**Conclustion**: I have been able to extract meaning from EmpJobLevel feature, Job level means position of the employee in the companys hierarchy tree. Since we will be knowing for what Job level a candidate will be interviewed for, this feature will be considered for ML modelling that will be used for employee hiring.

```python
#EmpJobLevel
df.EmpJobLevel.value_counts()
```

```python
jobLevelAnalysisDf = df[[
 'EducationBackground',
 'EmpDepartment',
 'EmpJobRole',
 'BusinessTravelFrequency',
 'EmpEducationLevel',
 'EmpHourlyRate',
 'EmpJobLevel',
 'OverTime',
 'EmpRelationshipSatisfaction',
 'TotalWorkExperienceInYears',
 'TrainingTimesLastYear',
 'EmpWorkLifeBalance',
 'ExperienceYearsAtThisCompany',
 'ExperienceYearsInCurrentRole',
 'YearsSinceLastPromotion',
 'YearsWithCurrManager',
 'PerformanceRating']]
#These features seems to be correlated with EmpJobLevel(assuming job level means position in Companys hierarchy)
```

```python
from sklearn.preprocessing import LabelEncoder
enc = LabelEncoder()
for labelColumn in ['EmpDepartment', 'EmpJobRole', 'BusinessTravelFrequency', 'OverTime', ]:
    jobLevelAnalysisDf[labelColumn] = enc.fit_transform(jobLevelAnalysisDf[labelColumn])
```

```python
manyFeatureCorrDf = jobLevelAnalysisDf.corr()
top_corr_features = manyFeatureCorrDf.index
plt.figure(figsize=(12,12))
#plot heat map
g=sns.heatmap(jobLevelAnalysisDf[top_corr_features].corr(),annot=True,cmap="RdYlGn")
```

```python
absoluteCorrelationOfEmployeeJobLevelWithRest = manyFeatureCorrDf.EmpJobLevel.apply(lambda x: abs(x))

absoluteCorrelationOfEmployeeJobLevelWithRest.sort_values(ascending=False)
```

### Analysis of feature 'overtime'

What is the meaning of overtime in this context and how relevant is it for Employee hiring?  

Assumption 1: overtime=Yes means Employee is ok to work overtime  
>If this is true, employees willing to work overtime will have significantly high hourly rate - From below visuals this assumption is WRONG    
>If this is true, employees willing to work overtime will have significantly low 'Years Since Last Promotion'  - From below visuals this assumption is WRONG    
  
  
Assumption 2: overtime=Yes means Employee is in a job role where overtime is required  
>If this is true, only certain job roles should have employees with most overtime=yes employee. - From below correlation analysis overtime doesn't seem to have good correlation either with Job level or with Department.  

**Conclusion**: I have not been able to establish meaning of the feature 'overtime' and hence not considering for job hiring ML model.  

```python
#EmpHourlyRate
fig, ax = plt.subplots(figsize=(6, 6))
overTimeList = df.groupby('OverTime').mean().EmpJobInvolvement.sort_values(ascending=False).index.tolist()

sns.barplot(x="OverTime", y="EmpHourlyRate", 
            ax=ax, 
            order=overTimeList,
            data=df);

ax.set_title('Employee hourly rate based on willingness to work overtime', fontdict={'fontsize' : 20});
ax.set_xlabel('Over Time', fontdict={'fontsize' : 20});
ax.set_ylabel('Employee Hourly Rate', fontdict={'fontsize' : 20});
ax.set_ylim(30, 75)

dispAvgLabelOnBarTop(ax);
```
