```python
#Utility/Useful functions

def dispAvgLabelOnBarTop(ax):
    for p in ax.patches:
        ax.annotate(format(p.get_height(), '.2f'), 
                (p.get_x() + ((p.get_width() - 0.5) / 2.), p.get_height()), 
                ha = 'center', va = 'center', 
                xytext = (0, 10), 
                textcoords = 'offset points')


        
#Utility functions
def reserveColorsAsPerCatFeatureOrder(ax, CatFeature, outColorDict):
    catFeatureList = df[CatFeature].value_counts().index.to_list()
    for i, c in enumerate(ax._get_lines.prop_cycler):
        if i >= len(catFeatureList):
            break
        #Update outColorDict    
        outColorDict[catFeatureList[i]] = c['color']



def dispEmpCountPercent(ax, CatFeature, DispOrderList):    
    
    CatFeatureCountDict = df[CatFeature].value_counts().to_dict()
    TotalEmployees = df[CatFeature].value_counts().sum()

    i=0
    for p in ax.patches:
        ax.annotate(format((100 * CatFeatureCountDict[DispOrderList[i]])/TotalEmployees, 
                           '.2f')+'%\nEmployees',
                (p.get_x() + ((p.get_width()) / 2.), p.get_height()/2.), 
                ha = 'center', va = 'center', 
                xytext = (0, 10), 
                textcoords = 'offset points',
               )
        i += 1;
        
```

```python
fig, ax = plt.subplots(figsize=(14, 8))
deptList = df.groupby('TrainingTimesLastYear').mean().PerformanceRating.sort_values(ascending=False).index.tolist()
sns.barplot(x="TrainingTimesLastYear", y="PerformanceRating", 
            ax=ax, 
            order=deptList,
            data=df);

    
ax.set_title('Average   Performance   rating  in  each   of Training Times ', fontdict={'fontsize' : 20});
ax.set_ylabel('Performance Rating', fontdict={'fontsize' : 20});
ax.set_xlabel('TrainingTimesLastYear', fontdict={'fontsize' : 20});


dispAvgLabelOnBarTop(ax);

dispEmpCountPercent(ax, 'TrainingTimesLastYear', deptList);
```

**Insight 9:**   
About 70% of the employees received 2 or 3 trainings last year. There is no significant difference in performance among employees who have taken various number of trainings ranging between 0 to 6 times

## Section 2: Identifying Top 3 important features effecting employee rating

**Note :**  
Using correlation heatmap, we have already figured out top 3 important features effecting employee rating.  
We can train a ML model with all given features taking employee performance rating target variable. We can let ML algorithms figure out the importance of each feature with respect employee performance rating

```python
X = df.drop('PerformanceRating', axis=1)

y = df.PerformanceRating
```

```python
#Label Encode categorical-label columns
from sklearn.preprocessing import LabelEncoder
enc = LabelEncoder()
for labelColumn in cat_columns:
    X[labelColumn] = enc.fit_transform(X[labelColumn])
```

```python
X.head(2)
```

```python
#No need of splitting as I am trying to find out all important features

from xgboost import XGBClassifier
modelxg = XGBClassifier(learning_rate=0.1, random_state=0, n_estimators=200)

modelxg.fit(X, y)
```

```python
FeatureImportance = pd.DataFrame({'Feature':X.columns,'Importance':modelxg.feature_importances_*100})

FeatureImportance = FeatureImportance.sort_values(by='Importance', axis=0, ascending=False)


FeatureImportance.plot(x='Feature',
            y='Importance',
            kind='bar',
            grid=True,
            title="Feature Vs Influence on Model",
            #colormap='BrBG',
            legend = True,
            figsize = (16,8)
            );
```

```python
print('Top 5 Features are : ')
for i,feature in enumerate(FeatureImportance.Feature.values.tolist()[0:5]):
    print(i+1,' : ', feature)
```

**conclusion 2:**  
Using XGBoost ML algorithm, we figured out that the top 3 features affecting employee performance are    
same as those features found out by correlation

**FINAL CONCLUSION :**
